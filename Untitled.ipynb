```python
import cv2
from ultralytics import YOLO
import os
# YOLOv8 모델 로드 (사전 훈련된 모델 사용)
model = YOLO("runs/detect/train2/weights/best.pt")  
# 학습된 모델 경로 설정, epochs=50, imgsz=640, device=device)  
# 데이터셋 경로, 에포크 수, 이미지 크기 설정

# 동영상 파일 경로 설정
input_video_path = 'objectdetection.mp4'  # 입력 동영상 파일 경로
output_video_path = 'output_video.mp4'  # 출력 동영상 파일 경로
# 클래스 필터 (Book, Chair, Mouse)
class_names = ['book', 'keyboard', 'mouse']
class_ids = [k for k, v in model.names.items() if v in class_names]

# 동영상 캡처 객체 생성
cap = cv2.VideoCapture(input_video_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# 출력 동영상 객체 생성
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # YOLOv8 모델을 사용하여 객체 탐지 수행
    results = model(frame)
    
    # 결과 필터링 및 렌더링
    for result in results:
        boxes = result.boxes.cpu().numpy()
        for box in boxes:
            cls_id = int(box.cls[0])
            if cls_id in class_ids:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                conf = box.conf[0]
                label = f"{model.names[cls_id]} {conf:.2f}"
                
                # 바운딩 박스 그리기
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
    
    # 렌더링된 프레임을 출력 동영상에 작성
    out.write(frame)

# 리소스 해제
cap.release()
out.release()

print(f"동영상 저장 완료: {output_video_path}")
```
